```python
import json
from hanziconv import HanziConv
```

```python
pos_path = 'extern/Chinese-Sentiment-Lexicon/positive_submit.txt'
neg_path = 'extern/Chinese-Sentiment-Lexicon/negative_submit.txt'
out_path = 'lexicon_mixed.json'
inverted_filename = 'inverted_file.json'
```

```python
MAXGRAM = 3
```

```python
# load vocab
inverted = json.load(open(inverted_filename, 'r'))
vocab = {}
for (word, voc) in inverted.items():
    ngram = len(word)
    if ngram > MAXGRAM:
        continue
    i = len(vocab)
    vocab[word] = i

VOC_SZ = len(vocab)
print(VOC_SZ)
```

```python
extern = {'positive':[], 'negative':[]}
```

```python
# positive
with open(pos_path, 'r') as f:
    for line in f:
        trad = HanziConv.toTraditional(line.strip())
        if trad in vocab:
            extern['positive'].append(trad)
```

```python
# negative
with open(neg_path, 'r') as f:
    for line in f:
        trad = HanziConv.toTraditional(line.strip())
        if trad in vocab:
            extern['negative'].append(trad)
```

```python
querylexicon = {\
'negative': \
['反對',  '拒絕',  '不應',  '錯誤',  '不可', '不贊同', '不贊成', '不合理' ],\
'positive': \
['支持', '同意', '應該', '正確', '可以', '贊同', '贊成', '合理', '合法', '應', '可', '有']}
#not found '不對',   '不支持',     '不同意', '不應該', '不正確','不可以','不合法','不反對', '對的', 
```

```python
comb_lexi = {'positive':list(set(querylexicon['positive']+extern['positive'])),\
             'negative':list(set(querylexicon['negative']+extern['negative']))}
assert all([w in comb_lexi['positive'] for w in querylexicon['positive']+extern['positive']])
assert all([w in comb_lexi['negative'] for w in querylexicon['negative']+extern['negative']])
print(comb_lexi)
```

```python
json.dump(comb_lexi, open(out_path, 'w'))
```
